```python
# ── ЯЧЕЙКА 2 — ngrok токен ────────────────────────────
# Получить бесплатно: https://dashboard.ngrok.com/get-started/your-authtoken

NGROK_TOKEN = 'YOUR_TOKEN_HERE'  # ← вставь сюда

from pyngrok import ngrok
ngrok.set_auth_token(NGROK_TOKEN)
print('✅ Токен установлен')
```

```python
# ── ЯЧЕЙКА 3 — Загрузка модели ───────────────────────
# Занимает 5–10 минут. Требует GPU.

import torch
from transformers import Qwen2_5_VLForConditionalGeneration, AutoProcessor
from PIL import Image

MODEL_ID = 'Qwen/Qwen2.5-VL-3B-Instruct'

print('⏳ Загружаю модель...')

processor = AutoProcessor.from_pretrained(MODEL_ID)
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    device_map='auto'
)
model.eval()

print('✅ Модель загружена!')
print('GPU памяти занято:', round(torch.cuda.memory_allocated() / 1e9, 1), 'GB')
```
